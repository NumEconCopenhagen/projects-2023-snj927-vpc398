## Cobb Douglas production model
In this assignment, we will try compute output from a Cobb Douglas production function given different inputs and parameter values.

**Theory behind the CD production model**

The Cobb Douglas production function used in this assignment is stated below. It describes a standard form for production of a single good with two factors:
$$Y=AL^{\beta}K^{\alpha}$$

$Y$ denotes the total production (all goods produces). $L$ and $K$ denotes respectively labour and capital input. Labour input describes the amount of hours worked and capital input is a measure of assets in terms of machinery, equipment, building, etc. $A$ denotes the total factor productivity, and the variable can be considered an efficiency parameter. 

The elasticities in the model is $\alpha$ and $\beta$ for the two inputs respectively. These values are determined by available technology why they are considered constants. We will look into the elasticities in this assignment, as they measure the responsiveness of output to a change in levels of the inputs used in production:
- If $\alpha + \beta = 1$, then the productions function will display constant return to scale (CRS)
- If $\alpha + \beta < 1$, then the productions function will display decreasing return to scale (DRS)
- If $\alpha + \beta > 1$, then the productions function will display increasing return to scale (IRS)

**Modelling the CD production function**

Below, the Cobb Douglas function is defined and tested by giving random values:

```python
# Defining the Cobb Douglas function and giving the elasticities random values
def cobb_douglas(K, L):
    A = 1.5
    beta = 1
    alpha = 0.33
    # The sum of elasticities above (alpha and beta) indicate IRS in this case
    return A * L**beta * K**alpha

# Simulating with IRS, A = 1.0, K = 1,5, and L = 2.0
print("The production output given the values of input and elasticities is:")
print("Y =", cobb_douglas(1.5, 2.0))
```

**Multiple returns**

We will now try to analyze the CD production function and how the output changes as we change one input through marginal product. We will start by defining marginal product of labor (MPL) and marginal product of capital (MPK):

$$MPL(K,L)=\partial Y(K,L) / \partial L$$
$$MPK(K,L)=\partial Y(K,L) / \partial K$$

Both marginal products will now be calculated and then return both:

```python
# Define marginal products:
def marginal_products(K, L, epsilon):
    MPL = (cobb_douglas(K, L + epsilon) - cobb_douglas(K, L)) / epsilon
    MPK = (cobb_douglas(K + epsilon, L) - cobb_douglas(K, L)) / epsilon
    return MPL, MPK

# Using tuple to compute MPL and MPK
tuple=marginal_products(1.0, 0.5, 1e-4)
print(tuple)
```

We can use this to calculate the marginal products for different values of capital input, fixing L using a comprehension:

```python
# HVAD GØR VI HER CODING WISE?
Ks = [2.0, 4.0, 6.0]
[marginal_products(K, 0.5, 1e-4) for K in Ks]
```

**Simulation and illustrating the Cobb Douglas production function**

In the first simulation, we would like to investige the relationship between $K$ and $L$ in relation to the output level, $Y$:

```python
# import magic
import numpy as np
import matplotlib.pyplot as plt

# Define the Cobb-Douglas function
def cobb_douglas(K, L):
    A = 1.5
    beta = 1
    alpha = 0.33
    return A * L**beta * K**alpha

# Range of values for K and L
K_values = np.linspace(0.1, 10, 100)
L_values = np.linspace(0.1, 10, 100)

# Calculate the output (Y) for each combination of K and L
Y_values = np.zeros((len(K_values), len(L_values)))
for i, K in enumerate(K_values):
    for j, L in enumerate(L_values):
        Y_values[i, j] = cobb_douglas(K, L)

# Create a 3D plot of the production function
K, L = np.meshgrid(K_values, L_values)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(K, L, Y_values)
ax.set_xlabel('Capital (K)')
ax.set_ylabel('Labor (L)')
ax.set_zlabel('Output (Y)')
ax.set_title('Cobb-Douglas Production Function')

# Show the plot
plt.show()
```

**Kommenter ovenstående og analysér (kort)**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Cobb-Douglas function
def cobb_douglas(K, L, alpha, beta):
    A = 1.5
    return A * (L**beta) * (K**alpha)

# Range of values for K and L
K_values = np.linspace(0.1, 10, 100)
L_values = np.linspace(0.1, 10, 100)

# Parameters for different scenarios
return_to_scale_values = [0.8, 1.0, 1.2]
A = 1.5
K = 5.0
L = 5.0

# Calculate the output (Y) range across all scenarios
Y_min = float('inf')
Y_max = float('-inf')
for return_to_scale in return_to_scale_values:
    alpha = 0.5 * return_to_scale
    beta = 0.5 * return_to_scale
    Y = cobb_douglas(K, L, alpha, beta)
    Y_min = min(Y_min, Y)
    Y_max = max(Y_max, Y)

# Plot the production function for each scenario
fig, axes = plt.subplots(1, len(return_to_scale_values), figsize=(18, 6), sharex=True, sharey=True)

for i, return_to_scale in enumerate(return_to_scale_values):
    alpha = 0.5 * return_to_scale
    beta = 0.5 * return_to_scale

    # Calculate the output (Y) for each combination of K and L
    Y_values = np.zeros((len(K_values), len(L_values)))
    for j, K in enumerate(K_values):
        for k, L in enumerate(L_values):
            Y_values[j, k] = cobb_douglas(K, L, alpha, beta)

    # Plot the production function
    ax = axes[i]
    c = ax.contourf(K_values, L_values, Y_values, levels=20, cmap='viridis', vmin=Y_min, vmax=Y_max)
    ax.set_xlabel('Capital (K)')
    ax.set_ylabel('Labor (L)')
    ax.set_title(f'Return to Scale: {return_to_scale} (A = {A})')
    fig.colorbar(c, ax=ax, label='Output (Y)')

plt.tight_layout()
plt.show()
```

The Cobb-Douglas production function is shown graphically, showing the relationship between the inputs of labor and capital and the resulting output or GDP. The horizontal axis of the graph represents labor input and the vertical axis represents capital input. Output (GDP) is represented by a series of isolines (production curves) representing different levels of output that can be produced with different combinations of labor and capital inputs. The MRTS, or marginal technology substitution rate, is represented by the slope of each isocant and describes the rate at which a firm can switch from one input to another while maintaining the same level of output. Moving left to right along the isolines flattens the slope, indicating that more capital is needed to offset a small decrease in work, and vice versa. At the intersection of the isolines and the isocost lines, the best mix of labor and capital inputs that maximize output or profit can be found. The slope of the isocost line is determined by the ratio of labor and capital prices.
